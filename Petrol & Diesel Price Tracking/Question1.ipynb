```python
#import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

#1. Data Loading and Initial Display
df = pd.read_csv('petrolprices.csv', delimiter = ';')

print("Initial Data Preview: ")
print(df.head())

#2. Data Processing
#set year and month to index
df['Date'] = df['Year'].astype(str) + '-' + df['Month']
df.set_index('Date', inplace = True)

#Create price difference columm
df['Price Difference'] = df['Diesel'] - df['Petrol']

#3. Yearwise- Data segementation
df_2023 = df[df.index.str.contains('2023')]
df_2024 = df[df.index.str.contains('2024')]

#4. Data visualisation
#Individual Line Charts for prteol and diesel prices

#2023
plt.figure(figsize = (10,5))
plt.plot(df_2023.index, df_2023['Petrol'], label = 'Petrol 2023')
plt.plot(df_2023.index, df_2023['Diesel'], label = 'Diesel 2023')
plt.title("Petrol and Diesel Prices - 2023")
plt.xlabel("Month")
plt.ylabel("Price (ZAR)")
plt.legend()
plt.tight_layout()
plt.xticks(rotation = 90)
plt.savefig("2023_LineChart.png")
plt.show()

#2024
plt.figure(figsize = (10,5))
plt.plot(df_2024.index, df_2024["Petrol"], label = "Petrol 2024", color = "red")
plt.plot(df_2024.index, df_2024["Diesel"], label = "Diesel 2024", color = "hotpink")
plt.title("Petrol and Diesel Prices -2024")
plt.xlabel("Month")
plt.ylabel("Price (ZAR)")
plt.xticks(rotation = 90)
plt.legend()
plt.tight_layout()
plt.savefig("2024_LineChart.png")
plt.show()

#Combined line chart from jan 2023 to dec 2024
plt.figure(figsize = (10,5))
plt.plot(df.index, df["Petrol"], label = "Petrol", color = "green")
plt.plot(df.index, df["Diesel"], label = "Diesel", color = "yellow")
plt.title("Petrol and Diesel Prices from January 2023 to December 2024")
plt.xlabel("Month")
plt.ylabel("Price(ZAR)")
plt.legend()
plt.xticks(rotation = 90)
plt.savefig("Jan2023_Dec2024.png")
plt.show()

#Price difference line chart
plt.figure(figsize = (10,5))
plt.plot(df.index, df["Price Difference"], label = "Price Difference", color = "purple")
plt.title("Price Difference Between Diesel and Petrol (2023-2024)")
plt.xlabel("Month")
plt.ylabel("Price Difference(ZAR)")
plt.legend()
plt.xticks(rotation = 90)
plt.savefig("price_difference(2023-2024).png")
plt.show()
```
